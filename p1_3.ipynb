```python
# Calculate and draw the path of a robot given a heading.  Draw also the
# shortest paths to the nearest horizontal and vertical walls.

import random
import math
import matplotlib.pyplot as plt 
  # Use plt as an alias so we don't have to type out matplolib.pyplot every time

# Set up the window
plt.close('all')        # close all currently open figure windows
plt.figure()            # creates a new figure window
ax= plt.gca()           # gets current axes
ax.set_aspect('equal')  # unit lengths on x- and y-axis are equal (equal scaling)

# Parameter
side= 10.    # side length of the square room

# Draw walls in blue, lower-left corner at (0,0)
plt.axis((-1, side+1, -1, side+1))  # x-axis limits -1 to side+1, 
                                    # y-axis limits -1 to side+1
plt.plot([0, side],    [0, 0], 'b-', linewidth=3)  # format: Blue line of width 3
plt.plot([side, side], [0, side], 'b-', linewidth=3)
plt.plot([side, 0],    [side, side], 'b-', linewidth=3)
plt.plot([0, 0],       [side, 0], 'b-', linewidth=3)

### Do not change the code above ###
####################################
### Modify the code below ##########

# Location of Robot
xr = random.random()*8+1
print(f'The x coordinate is: {xr}')
yr = random.random()*7+1
print(f'The x coordinate is: {yr}')
plt.plot(xr, yr, 'ro')  # format: Red circle
plt.text(xr, yr, '  Robot')

# User-clicked point
plt.title("Click away from the robot to set the robot's heading")
plt.show()  # show the plot
point= plt.ginput(1)  # Accepts one user click; note time-out of 30 seconds
xu, yu= point[0]      # Stores x-coord of click in variable xu,
                      #        y-coord of click in variable yu
plt.plot(xu, yu, 'm+')   # format: Magenta cross

#Finding the slope between clicked point and robot
if xu!=xr:
    slope = (yu-yr)/(xu-xr)
    
elif xu == xr:
    if yu>yr:
        slope = math.inf
    elif yr>yu:
        slope = -math.inf
        
print(f'The slope is: {slope}')

#Finding the slope between robot and corners
slope_ne = (10-yr)/(10-xr)
slope_nw = (10-yr)/(xr)
slope_sw = (yr)/(xr)
slope_se = (yr)/(10-xr)

#Finding the angle between robot and corners
angle_ne = math.atan((abs(slope_ne-0))/(1-slope_ne*0))
angle_nw = math.pi-math.atan((abs(slope_nw-0))/(1-slope_nw*0))
angle_sw = math.atan((abs(slope_sw-0))/(1-slope_sw*0))+math.pi
angle_se = math.pi*2 - math.atan((abs(slope_se-0))/(1-slope_se*0))

print(f'The angle with right top corner is: {angle_ne*180/math.pi} degrees')
print(f'The angle with left top corner is: {angle_nw*180/math.pi} degrees')
print(f'The angle with bottom left corner is: {angle_sw*180/math.pi} degrees')
print(f'The angle with bottom right corner is: {angle_se*180/math.pi} degrees')


#Finding the angle of heading: robot with clicked point
angle_heading = abs(math.atan(slope))
print(angle_heading)


if slope>0:
    if yu>yr:
        angle = (angle_heading)
    elif yu<yr:
        angle = (math.pi + abs(angle_heading))
        
elif slope<0:
    if yu>yr:
        angle = (math.pi - (angle_heading))
    elif yu<yr:
        angle = (math.pi*2 - (angle_heading))
        
print(f'The angle between robot and clicked point is: {angle*180/math.pi} degrees')
    
#Using the quadrant of heading to calculate robot's endpoint

if angle > angle_se:
    x_final = 10
    y_final = yr + (math.tan(angle)*(10-xr))
    
elif angle == angle_se:
        y_final = 0
        x_final = 10

elif angle > angle_sw:
        y_final = 0
        x_final = xr - (math.tan(math.pi/2-angle)*(yr))
        
elif angle == angle_sw:
        y_final = 0
        x_final = 0
        
elif angle > angle_nw:
        x_final = 0
        y_final = yr - (math.tan(angle-math.pi)*(xr))
        
elif angle == angle_nw:
        y_final = 10
        x_final = 0
        
elif angle > angle_ne:
        y_final = 10
        x_final = xr + ((10-yr)/(math.tan(angle)))
        
elif angle == angle_ne:
        y_final = 10
        x_final = 10
    

else:
    y_final = yr + (10-xr)*(math.tan(angle))
    x_final = 10
    
    
print (x_final)
print (y_final)

plt.title(f"The robot's heading is {angle:.1f} radians")

plt.plot(x_final,y_final, 'c.')

# Draw line to connect the two points
plt.plot([xr, x_final], [yr, y_final], 'k-')  # format: blacK dotted line of default width

if xr<5:
    closest_x = 0
else: 
    closest_x = 10
    
if yr<5:
    closest_y = 0
else: 
    closest_y = 10
    
    
plt.plot([xr, closest_x], [yr, yr], 'm:')
plt.plot([xr, xr], [yr, closest_y], 'c--')


####################################
### Do not change the code below ###
print('Bye')
```
